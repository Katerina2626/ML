```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/разработка нейронных сетей/Mall_Customers.csv')
```

```python
data
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='Spending Score (1-100)', data=data)
plt.title('Box plot')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Genre', y='Annual Income (k$)', data=data)
plt.title('Violin plot of Annual Income by Gender')
plt.show()
```

```python
# Count plots
sns.countplot(x='Genre', data=data)
```

```python
# Heatmap
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
```

```python
# Facet grid
g = sns.FacetGrid(data, col='Genre', row='Age')
g.map(sns.scatterplot, 'Annual Income (k$)', 'Spending Score (1-100)')
```

```python
# Regplot
sns.regplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data)
```

```python
# lmplot
sns.lmplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=data, hue='Genre')
```

```python
# Swarm plot
sns.swarmplot(x='Genre', y='Annual Income (k$)', data=data)
```

```python
# Pair plot
sns.pairplot(data, hue='Genre')
```

```python
# Joint plot
sns.jointplot(x='Age', y='Annual Income (k$)', data=data, kind='hex')
```
