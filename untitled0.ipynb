```python
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
import pandas as pd
from matplotlib import pyplot
from scipy.stats import shapiro

seed(1)
data =pd.read_excel('/content/drive/MyDrive/Colab Notebooks/разработка нейронных сетей/олимпиады 2.2/датасет для 2.2.xlsx')
data4= data[["Gross in rubles (₽)[225]"]]
stat, p = shapiro(data4)
print('Statistics=%.3f, p=%.3f' % (stat,p))
alpha=0.05
if p > alpha:
  print("Sample looks Gaussian")
else:
  print("Sample does not look Gaussian")
```

```python
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
import pandas as pd
from matplotlib import pyplot
from scipy.stats import normaltest

seed(1)
data =pd.read_excel('/content/drive/MyDrive/Colab Notebooks/разработка нейронных сетей/олимпиады 2.2/датасет для 2.2.xlsx')
data5= data[["Gross in rubles (₽)[225]"]]

stat, p = normaltest(data5)
print("Statistics=%.3f, p=%.3f" % (stat,p))
alpha = 0.05
if p>alpha:
  print("Sample looks Gaussian")
else:
  print("Sample does not look Gaussian")

```

```python
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
import pandas as pd
from matplotlib import pyplot
from scipy.stats import normaltest
from scipy.stats import anderson

seed(1)
data =pd.read_excel('/content/drive/MyDrive/Colab Notebooks/разработка нейронных сетей/олимпиады 2.2/датасет для 2.2.xlsx')
data5= data[["Gross in rubles (₽)[225]"]]
result = anderson(data5)
print("Statistic: %.3f" % result.statistic)
p=0
for i in range(len(result.critical_values)):
  sl, cv = result.significance_level[i], result.critical_values[i]
  if result.statistic < result.critical_values[i]:
    print("%.3f: %.3f, data looks normal" %(sl,cv))
  else:
    print("%.3f: %.3f, data does looks normal" %(sl,cv))
```
